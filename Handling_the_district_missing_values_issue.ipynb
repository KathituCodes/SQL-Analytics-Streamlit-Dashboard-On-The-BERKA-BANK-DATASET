# Documentation

The district data had some issue importing.

## **Issue** :
Some cells in the district CSV contain missing data instead of a number. PostgreSQL does not understand NaN as a number so it throws an error.
We are going to fix this using Python. I uploaded the dataset and intend to replace every NaN with the mean, and save a clean version. Then we import the clean version.

```python
import numpy as np
import pandas as pd
```

```python
# Read the district CSV
df = pd.read_csv('/content/district.csv', sep=';', na_values=['?'])

# Load the head of the dataset
df.head(10)
```

```python
# Check the shape of the data
print(df.shape)
```

```python
# Check for missing values in the dataset
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
# Display what the missing values are
print("\nMissing values:")
print(df[df.isnull().any(axis=1)])
```

```python
"""
Since the data missing is not transactional data and will likely not affect transcations made by clients in the
bank I decided to fill the missing values with the column average.
It is quick, easy, and preserves the overall mean of the district data without reducing the datasets shape.
"""
```

```python
# Fill missing values with the column average
df['A12'] = df['A12'].fillna(df['A12'].mean())
df['A15'] = df['A15'].fillna(df['A15'].mean())

# Print sucess message
print("Sucessfully filled missing values")
```

```python
# Verify no more missing values
print("Missing values after cleaning:")
print(df.isnull().sum())
```

```python
# Save the clean file
df.to_csv('/content/district_clean.csv', sep=';', index=False)
print("\nClean file saved successfully")
```
